```python
#‚úÖ STEP 3: Histogram ‚Äî Visualize distribution of values (e.g., marks, pixel intensity)
import numpy as np
import matplotlib.pyplot as plt

marks = np.random.normal(70, 10, 100)  # mean=70, std=10, 100 students
print(marks)

plt.hist(marks, bins=10, color='orange', edgecolor='black')
plt.title('Distribution of Marks')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.show()
```

```python
#‚úÖ STEP 4: Pie Chart ‚Äî Show proportion of categories (e.g., pass/fail ratio)
import matplotlib.pyplot as plt

labels = ['Passed', 'Failed']
sizes = [80, 20]  # e.g., 80 students passed, 20 failed
colors = ['#66b3ff', '#ff9999']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle
plt.title('Result Breakdown')
plt.show()
```


üéØ Real Use: Useful in dashboards to quickly show proportions (e.g., class balance, sentiment analysis).


```python
#‚úÖ STEP 5: Scatter Plot ‚Äî Show relation between two continuous variables
# E.g., visualize the relationship between study time and marks.
import matplotlib.pyplot as plt

# Sample data
study_hours = [1, 2, 3, 4, 5, 6, 7, 8]
marks = [40, 50, 55, 65, 70, 75, 80, 90]

plt.scatter(study_hours, marks, color='purple') #marker parameter
plt.title('Study Time vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.grid(True)
plt.show()
```

```python
#‚úÖ STEP 8: Box Plot ‚Äî Visualize data distribution and detect outliers
#Example: Box plot of students‚Äô Marks
import matplotlib.pyplot as plt

marks = [85, 90, 70, 60, 95, 80, 100, 55, 78, 88]

plt.boxplot(marks)
plt.title('Box plot of Marks')
plt.ylabel('Marks')
plt.show()
```

üéØ Real Use: Quickly see median, quartiles, and identify outliers in data.

Median: The line inside the box is the median (50th percentile).

Quartiles:
The bottom edge of the box is the 25th percentile (Q1).
The top edge of the box is the 75th percentile (Q3).

Interquartile Range (IQR): Q3 - Q1, the height of the box.

Whiskers: Lines extending from the box to show the range of the data, usually up to 1.5 * IQR beyond Q1 and Q3.

Outliers: Points plotted beyond the whiskers (individual dots), indicating values significantly different from the rest.

```python
import numpy as np

marks = np.array([85, 90, 70, 60, 95, 80, 100, 55, 78, 88])

Q1 = np.percentile(marks, 25)
median = np.median(marks)
Q3 = np.percentile(marks, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Q1 (25th percentile):", Q1)
print("Median (50th percentile):", median)
print("Q3 (75th percentile):", Q3)
print("Interquartile Range (IQR):", IQR)
print("Lower Bound (for outliers):", lower_bound)
print("Upper Bound (for outliers):", upper_bound)

outliers = marks[(marks < lower_bound) | (marks > upper_bound)]
print("Outliers:", outliers)
```

üé® What is Seaborn?
Seaborn is a wrapper over Matplotlib that gives:

Simpler syntax for complex plots

Works great with Pandas DataFrames

```python
# üî• 2. sns.heatmap() ‚Äì Seaborn (built on top of Matplotlib)
# A prettier, annotation-ready version of imshow().
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Correlation matrix example
df = pd.DataFrame(np.random.rand(5, 5), columns=list('ABCDE'))
print(df)

plt.figure(figsize=(6, 5))
sns.heatmap(df, annot=True, cmap="YlGnBu", linewidths=0.5)
plt.title("Seaborn Heatmap with Annotations")
plt.show()
```

‚úÖ Final Step: Mini Project ‚Äî "Student Performance Visualizer" üìäüéì

```python
#üìÅ STEP 1: Create the dataset
import pandas as pd
import numpy as np

np.random.seed(42)

names = ['Arjun', 'Sneha', 'Ravi', 'Fatima', 'Krishna', 'Zara', 'Manoj', 'Lata', 'Vikram', 'Sara']
marks = np.random.randint(40, 100, size=10)
attendance = np.random.randint(50, 100, size=10)

df = pd.DataFrame({
    'Name': names,
    'Marks': marks,
    'Attendance': attendance
})
print(df)
```

üìä STEP 2: Data Visualizations
Bar Chart ‚Äì Student vs Marks

Pie Chart ‚Äì Grade Distribution

Line Chart ‚Äì Attendance trend

Scatter Plot ‚Äì Marks vs Attendance

```python
import matplotlib.pyplot as plt

# 1. Bar Chart
plt.figure(figsize=(8, 4))
plt.bar(df['Name'], df['Marks'], color='skyblue')
plt.title('Marks by Student')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Add Grades Column
def get_grade(m):
    if m >= 90:
        return 'A'
    elif m >= 75:
        return 'B'
    elif m >= 60:
        return 'C'
    else:
        return 'Fail'

df['Grade'] = df['Marks'].apply(get_grade)

# 3. Pie Chart
grade_counts = df['Grade'].value_counts()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', colors=['green','blue','orange','red'])
plt.title('Grade Distribution')
plt.axis('equal')
plt.show()

# 4. Line Chart (Attendance)
plt.plot(df['Name'], df['Attendance'], marker='o', linestyle='--', color='purple')
plt.title('Attendance Trend')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 5. Scatter Plot (Marks vs Attendance)
plt.scatter(df['Marks'], df['Attendance'], c='red')
plt.xlabel('Marks')
plt.ylabel('Attendance')
plt.title('Marks vs Attendance')
plt.grid(True)
plt.show()
```

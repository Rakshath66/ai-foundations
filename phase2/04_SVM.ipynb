```python
# ðŸ§  Step 1: Visualize SVM Margin Intuition
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Generate sample data
X, y = make_blobs(n_samples=100, centers=2, random_state=6)
# generates fake cluster data with 2 distict classes to test classifiers
# X- coordinates of points (shape(100,2)), y- class labels (0 or 1)
print("coordinates",X,end="\n")
print("class labels",y,end="\n")

# Train an SVM
clf = SVC(kernel='linear') #Tells the SVM to find a straight-line decision boundary (hyperplane)
clf.fit(X, y) # alculates the best hyperplane that separates the two classes with the maximum margin
# Other kernal options
 # - 'rbf' - for curves (Gaussian)
 # - 'poly' - for polynomial boundaries
 # - 'sigmoid' - for neural-like curves

# Plotting - plot on 2d space
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap='bwr') #c=y colours red/blue. cmap='bwr' blue= class0, red=class1

# Plot decision boundary
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Create grid
import numpy as np
xx = np.linspace(xlim[0], xlim[1])
yy = np.linspace(ylim[0], ylim[1])
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# Plot decision boundary and margins
plt.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], linestyles=['--','-','--'])

# Plot support vectors
plt.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
            linewidth=1, facecolors='none', edgecolors='k')

plt.title("SVM Margin & Support Vectors")
plt.show()
```
